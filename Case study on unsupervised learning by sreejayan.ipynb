# the optimal number of clusters for Agglomerative Clustering

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Set the range of cluster numbers to try
min_clusters = 2
max_clusters = 10

# Initialize empty lists to store evaluation metrics
kmeans_wcss = []
kmeans_silhouette_scores = []
agg_silhouette_scores = []

# Iterate through different numbers of clusters
for num_clusters in range(min_clusters, max_clusters+1):
    # K-means clustering
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(features)
    kmeans_wcss.append(kmeans.inertia_)
    kmeans_silhouette_scores.append(silhouette_score(features, kmeans.labels_))

    # Agglomerative Clustering
    agg_clustering = AgglomerativeClustering(n_clusters=num_clusters)
    agg_clustering.fit(features)
    agg_silhouette_scores.append(silhouette_score(features, agg_clustering.labels_))

# Plot the evaluation metrics
plt.figure(figsize=(12, 4))

# Elbow plot for K-means
plt.subplot(1, 2, 1)
plt.plot(range(min_clusters, max_clusters+1), kmeans_wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Elbow Curve')

# Silhouette score plots
plt.subplot(1, 2, 2)
plt.plot(range(min_clusters, max_clusters+1), kmeans_silhouette_scores, marker='o', label='K-means')
plt.plot(range(min_clusters, max_clusters+1), agg_silhouette_scores, marker='o', label='Agglomerative')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Comparison')
plt.legend()

plt.tight_layout()
plt.show()

# Find the optimum number of clusters
optimal_kmeans_clusters = kmeans_silhouette_scores.index(max(kmeans_silhouette_scores)) + min_clusters
optimal_agg_clusters = agg_silhouette_scores.index(max(agg_silhouette_scores)) + min_clusters

print("Optimum number of clusters (K-means):", optimal_kmeans_clusters)
print("Optimum number of clusters (Agglomerative):", optimal_agg_clusters)

# Create the models with the optimum number of clusters
kmeans_optimal = KMeans(n_clusters=optimal_kmeans_clusters, random_state=42)
kmeans_optimal.fit(features)

agg_clustering_optimal = AgglomerativeClustering(n_clusters=optimal_agg_clusters)
agg_clustering_optimal.fit(features)
```
