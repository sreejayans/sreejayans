```python
# Visualizations
# Distribution of Age
plt.figure(figsize=(8, 6))
sns.distplot(data['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(16,7))

plt.subplot(2,2,1)
sns.distplot(data['age'],color='DeepPink')

plt.show()
```

```python
# Distribution of Gender
plt.figure(figsize=(6, 4))
sns.countplot(data['gender'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Countries
plt.figure(figsize=(15, 6))
sns.countplot(data['country'], order=data['country'].value_counts().index)
plt.title('Distribution of Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation Heatmap
corr_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot histograms for numerical variables
data.hist(bins=10, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Select numerical attributes for the pair plot
numerical_attributes = [
    'age', 'ball_controlling_skills', 'body_reflexes', 'body_balance',
    'height', 'weight', 'behaviour_rating', 'matches_played',
    'fitness_rating', 'trophies_won', 'dedication_level',
    'coaching', 'years_of_experience', 'no_of_disqualifications'
]

# Create a pair plot
sns.set(style="ticks")
sns.pairplot(data[numerical_attributes])
plt.suptitle("Pair Plot of Football Player Attributes", y=1.02)
plt.show()
```

```python
# Check the distribution of the target class
plt.figure(figsize=(6, 4))
sns.countplot(data['selection'])
plt.title('Distribution of Selection (Target Class)')
plt.xlabel('Selection')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize numerical features by target class
numerical_attributes = [
    'age', 'ball_controlling_skills', 'body_reflexes', 'body_balance',
    'behaviour_rating', 'matches_played',
    'fitness_rating', 'trophies_won', 
     'years_of_experience'
]
for attribute in numerical_attributes:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='selection', y=attribute, data=data)
    plt.title(f'{attribute} by Selection')
    plt.xlabel('Selection')
    plt.ylabel(attribute)
    plt.show()
```

```python
# Visualize categorical features by target class
categorical_attributes = ['gender', 'country', 'strong_foot']

for attribute in categorical_attributes:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=attribute, hue='selection', data=data)
    plt.title(f'{attribute} by Selection')
    plt.xlabel(attribute)
    plt.ylabel('Count')
    plt.legend(title='Selection')
    plt.xticks(rotation=45)
    plt.show()
```
