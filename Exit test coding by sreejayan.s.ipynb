# Missing value filling in dataset

```python

# Fill missing values in numeric columns with median
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median())



# Display the updated information about missing values
print(data.isnull().sum())

```

```python
#dropping unnecessary columns from the dataset
data.drop(['player_id', 'name', 'coaching', 'dedication_level', 'no_of_disqualifications'], axis = 1, inplace = True)
```

```python
from sklearn.impute import SimpleImputer

# Columns with missing values
columns_with_missing = ['gender', 'weight', 'strong_foot']

# Apply SimpleImputer to fill missing values
imputer = SimpleImputer(strategy='most_frequent')  # You can choose different strategies
data[columns_with_missing] = imputer.fit_transform(data[columns_with_missing])

```

```python
data.isnull().sum()
```

# Flooring And Capping are done in columns

```python
for column in [ 'age', 'ball_controlling_skills', 'body_reflexes', 'body_balance',
    'behaviour_rating', 'matches_played',
    'fitness_rating', 'trophies_won', 
     'years_of_experience'] :
    Q1 =data[column].quantile(0.25)
    Q3 =data[column].quantile(0.75)
    IQR=Q3 - Q1
    low=Q1 - (1.5*IQR)
    upr=Q3 + (1.5*IQR)
    data[column]=np.where(data[column] > upr, upr,np.where(data[column]< low,low,data[column]))
```

# Encoding

```python
# categorical columns
cat= [x for x in data.columns if data[x].dtypes=='O']

#Encoding
from sklearn.preprocessing import LabelEncoder
for x in cat:
    le=LabelEncoder()
    data[x]=le.fit_transform(data[x])
```

# model creation for predict selection of player

```python
x = data.drop('selection', axis=1)
y = data['selection']
```

```python
#splitting dataset into training and testing datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
logic = LogisticRegression()
logic.fit(x_train, y_train)
y_lr = logic.predict(x_test)

print(accuracy_score(y_test, y_lr))
print(confusion_matrix(y_test, y_lr))
```

# KNN- model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
metric = []
neighbors = np.arange(3,15)
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski', p = 2)
    classifier.fit(x_train, y_train)
    y_pred_knn = classifier.predict(x_test)
    acc = accuracy_score(y_test, y_pred_knn)
    metric.append(acc)


plt.plot(neighbors,metric,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
plt.show()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 14, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
y_knn = classifier.predict(x_test)

print('Accuracy =', accuracy_score(y_test, y_knn))
print(confusion_matrix(y_test, y_knn))
```

# Decision Tree model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_dt = dt.predict(x_test)

print('Accuracy =', accuracy_score(y_test, y_dt))
print(confusion_matrix(y_test, y_dt))
```

# RandomForestClassifier model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix


# Initialize the RandomForestClassifier
rf = RandomForestClassifier()

# Train the model on the training data
rf.fit(x_train, y_train)

# Predict using the trained model
y_rf = rf.predict(x_test)

# Calculate accuracy and print confusion matrix
accuracy = accuracy_score(y_test, y_rf)
confusion = confusion_matrix(y_test, y_rf)

print('Accuracy =', accuracy)
print('Confusion Matrix:')
print(confusion)
```

# Fine tuning 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV, train_test_split


# Define the hyperparameters you want to search through
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV with the RandomForestClassifier and the hyperparameter grid
grid_search = GridSearchCV(rf, param_grid, cv=3, n_jobs=-1)

# Perform grid search and find the best parameters
grid_search.fit(x_train, y_train)

# Get the best model after grid search
best_rf = grid_search.best_estimator_

# Predict using the best model
y_rf = best_rf.predict(x_test)


# Calculate accuracy and print confusion matrix
accuracy = accuracy_score(y_test, y_rf)
confusion = confusion_matrix(y_test, y_rf)

print('Best Parameters:', grid_search.best_params_)
print('Accuracy =', accuracy)
print('Confusion Matrix:')
print(confusion)
```
